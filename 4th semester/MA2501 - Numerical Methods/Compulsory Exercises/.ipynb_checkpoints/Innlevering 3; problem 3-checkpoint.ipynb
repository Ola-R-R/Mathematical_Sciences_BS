```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
```

```python
def trapz(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    h = float(b - a) / n
    fx = f(x[1:-1])
    ct = h * (np.sum(fx) + 0.5 * (f(x[0]) + f(x[-1])))
    return ct
```

```python
def simp(f, a, b, n):
    h = (b - a) / n
    first = f(a)
    last = f(b)
    
    x = a
    summ = 0
    for i in range(n - 1):
        x += h
        value = f(x)
        if i % 2 == 0:
            summ += 4 * value
        else:
            summ += 2 * value
    total = (h / 3) * (first + summ + last)
    return total
```

```python
def f1(x):
    return abs(x)**3

int_f = 1/2

a, b = -1, 1
errs = []
hs = []

for n in range(2, 100, 2):
    cqr_f = trapz(f1, a, b, n)
    # print("Number of subintervals n = {}".format(n))
    # print("Q[f] = {}".format(cqr_f))
    err = abs(int_f - cqr_f)
    errs.append(err)
    hs.append((b - a) / n)
    # print("I[f] - Q[f] = {:.10e}".format(err))
hs = np.array(hs)
errs = np.array(errs)
eoc = np.log(errs[1:]/errs[:-1])/np.log(hs[1:]/hs[:-1])
# print(eoc)
plt.title("f1 composite trapezoidal error rate")
plt.xlabel("Number of evaluation points", fontsize=13)
plt.ylabel("Absolute integration error", fontsize=13)
plt.loglog(hs, errs, "bo-", label = "|x|^3")
plt.legend()
plt.show()
```
