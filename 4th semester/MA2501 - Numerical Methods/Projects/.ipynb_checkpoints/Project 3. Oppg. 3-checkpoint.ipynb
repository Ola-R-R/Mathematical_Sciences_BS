```python
import numpy as np
from matplotlib import pyplot as plt
```

Modified Euler's method:
$$y_{n+1}=y_n+hf\left(x_n+\frac{1}{2}h,y_n+\frac{1}{2}hf(x_n,y_n)\right)$$

System of ODEs:
$$x'=10(y-x)$$
$$y'=x(28-z)-y$$
$$z'=xy-\frac{8}{3}z$$

Initial values:
$$x=1$$
$$y=z=0$$

Going to solve using step count of $100000$ and $100001$, $h\approx0.0005$

```python
# Defining function as a (3x1) vector:
def f(y):
    F_dotx = 10 * (y[1] - y[0])
    F_doty = y[0] * (28 - y[2]) - y[1]
    F_dotz = y[0] * y[1] - (8/3) * y[2]
    return np.array([F_dotx, F_doty, F_dotz]).reshape(3,1)

# Defining separate functions:
def f1(x, y, z):
    return 10 * (y - x)

def f2(x, y, z):
    return x * (28 - z) - y

def f3(x, y, z):
    return x * y - (8/3) * z

# Initial values:
yn = np.array([1,0,0]).reshape(3,1)

x = 1
y = 0
z = 0

# Step sizes:
h1 = 0.0005
h2 = 50 / 100001
```

```python
# Solving using vectors:
def mEV(f,y,h,it):
    y1 = []
    y2 = []
    y3 = []
    tt = np.arange(0,50,h)
    
    for i in range(it):
        # Adding values to make the graph:
        y1.append(y[0])
        y2.append(y[1])
        y3.append(y[2])
        
        # Step:
        y = y + h * f(y + 0.5 * h * f(y))
    
    # print(y)
    
    # Graphing
    fig, ax = plt.subplots(3, sharex=True, sharey=True)
    fig.suptitle("Solving using vectors with " + str(it) + " iterations")
    ax[0].plot(tt,y1)
    ax[0].set_title("x")
    ax[1].plot(tt,y2)
    ax[1].set_title("y")
    ax[2].plot(tt,y3)
    ax[2].set_title("z")
    fig.tight_layout()

# Solving with separate equations:
def mES(f1,f2,f3,x,y,z,h,it):
    xx = []
    yy = []
    zz = []
    tt = np.arange(0,50,h)
    
    for i in range(it):
        # Adding values to make the graph:
        xx.append(x)
        yy.append(y)
        zz.append(z)
        
        # f(yn + 0.5 * h * f(yn))
        p1 = f1(x + 0.5 * h * f1(x,y,z), y + 0.5 * h * f2(x,y,z), z + 0.5 * h * f3(x,y,z))
        p2 = f2(x + 0.5 * h * f1(x,y,z), y + 0.5 * h * f2(x,y,z), z + 0.5 * h * f3(x,y,z))
        p3 = f3(x + 0.5 * h * f1(x,y,z), y + 0.5 * h * f2(x,y,z), z + 0.5 * h * f3(x,y,z))
        
        # Step:
        x += h * p1
        y += h * p2
        z += h * p3 
    
    # print(x)
    # print(y)
    # print(z)
    
    fig, ax = plt.subplots(3, sharex=True, sharey=True)
    fig.suptitle("Solving separate equations with " + str(it) + " iterations")
    ax[0].plot(tt,xx)
    ax[0].set_title("x")
    ax[1].plot(tt,yy)
    ax[1].set_title("y")
    ax[2].plot(tt,zz)
    ax[2].set_title("z")
    fig.tight_layout()
```

```python
# 100000 step count:

mEV(f,yn,h1,100000)

mES(f1,f2,f3,x,y,z,h1,100000)
```

```python
# 100001 step count:

mEV(f,yn,h2,100001)

mES(f1,f2,f3,x,y,z,h2,100001)
```

```python
# Using initial values x = 1, y = 0.0001, z = 0:

mEV(f,np.array([1,0.0001,0]).reshape(3,1),h1,100000)

mES(f1,f2,f3,x,0.0001,z,h1,100000)
```
