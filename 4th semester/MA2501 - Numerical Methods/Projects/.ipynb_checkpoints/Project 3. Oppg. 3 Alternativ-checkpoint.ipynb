```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f1(x, y, z):
    return 10*(y-x)
def f2(x, y, z):
    return x*(28-z)-y
def f3(x, y, z):
    return x*y-(8/3)*z

x = 1
y = 0
z = 0
t = 0.0
h1 = 0.0005
h2 = 50 / 100001
```

```python
def mE1(f1,f2,f3,x,y,z,h):
    xx = []
    yy = []
    zz = []
    tt = np.arange(0,50,h)
    
    for i in range(100000):
        xx.append(x)
        yy.append(y)
        zz.append(z)
    
        x += h * f1(x + 0.5 * h * f1(x,y,z), y + 0.5 * h * f1(x,y,z), z + 0.5 * h * f1(x,y,z))
        y += h * f2(x + 0.5 * h * f2(x,y,z), y + 0.5 * h * f2(x,y,z), z + 0.5 * h * f2(x,y,z))
        z += h * f3(x + 0.5 * h * f3(x,y,z), y + 0.5 * h * f3(x,y,z), z + 0.5 * h * f3(x,y,z))
    
    print(x)
    print(y)
    print(z)
    
    plt.plot(tt,xx)
    plt.plot(tt,yy)
    plt.plot(tt,zz)

def mE2(f1,f2,f3,x,y,z,h):
    xx = []
    yy = []
    zz = []
    tt = np.arange(0,50,h)
    
    for i in range(100001):
        xx.append(x)
        yy.append(y)
        zz.append(z)
    
        x += h * f1(x + 0.5 * h * f1(x,y,z), y + 0.5 * h * f1(x,y,z), z + 0.5 * h * f1(x,y,z))
        y += h * f2(x + 0.5 * h * f2(x,y,z), y + 0.5 * h * f2(x,y,z), z + 0.5 * h * f2(x,y,z))
        z += h * f3(x + 0.5 * h * f3(x,y,z), y + 0.5 * h * f3(x,y,z), z + 0.5 * h * f3(x,y,z))
    
    print(x)
    print(y)
    print(z)
    
    plt.plot(tt,xx)
    plt.plot(tt,yy)
    plt.plot(tt,zz)
```
