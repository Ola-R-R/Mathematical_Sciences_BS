```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# initial value of t, y
k=1
t=0.0
y=5.0
y0=y
tt=[]
yy=[]
tf=10 # final value of x
dt=0.01 # step length

# defining function
def f(t,y,k):
    return -k*y

while t<=tf:
    tt.append(t)
    yy.append(y)
    k1=dt*f(t,y,k)
    k2=dt*f(t+dt,y+k1,k)
    y=y+0.5*(k1+k2)
    t+=dt

# Exact solution
tt=np.array(tt)
exact=y0*np.exp(-k*tt)
# Plotting of solution with exact result
plt.subplot(2,1,1)
plt.plot(tt,yy,'k--',label="dt=%.4f"%(dt))
plt.plot(tt, exact,'k',label="Exact solution")
plt.xlabel("time")
plt.ylabel("y")
plt.legend(loc='best')
plt.title("ODE by Modified Euler method")
# Calculation of Error and plotting
diff=exact-yy
plt.subplot(2,1,2)
plt.plot(tt,diff,'k.',label="Absolute error")
plt.xlabel("time")
plt.ylabel("Error")
plt.legend(loc='best')
plt.savefig("decay-m-euler-error.png")
```
